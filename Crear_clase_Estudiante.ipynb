```python
class Empleado:
    def __init__(self, nombre, apellido, cedula, email, semestre, materias):
        self.nombre = nombre
        self.apellido = apellido
        self.cedula = cedula
        self.email = email
        self.semestre = semestre
        self.materias = materias

    # Nombre
    @property
    def nombre(self):
        return self.__nombre

    @nombre.setter
    def nombre(self, value):
        self.__nombre = value

    # Apellido
    @property
    def apellido(self):
        return self.__apellido

    @apellido.setter
    def apellido(self, value):
        self.__apellido = value

    # Cédula (validación: 10 dígitos numéricos)
    @property
    def cedula(self):
        return self.__cedula

    @cedula.setter
    def cedula(self, value):
        value = str(value)
        if value.isdigit() and len(value) == 10:
            self.__cedula = value
        else:
            raise ValueError("La cédula debe tener exactamente 10 dígitos numéricos")

    # Email
    @property
    def email(self):
        return self.__email

    @email.setter
    def email(self, value):
        self.__email = value

    # Semestre (validación: entre 1 y 9)
    @property
    def semestre(self):
        return self.__semestre

    @semestre.setter
    def semestre(self, value):
        if isinstance(value, int) and 1 <= value <= 9:
            self.__semestre = value
        else:
            raise ValueError("El semestre debe ser un número entre 1 y 9")

    # Materias
    @property
    def materias(self):
        return self.__materias

    @materias.setter
    def materias(self, value):
        self.__materias = value


# Ejemplo de uso
empleado = Empleado(
    "Eylen",
    "Flores",
    "1234567890",
    "eylen@gmail.com",
    3,
    ["Programación", "Matemáticas"]
)

print(empleado.nombre)
print(empleado.cedula)
print(empleado.semestre)
```
