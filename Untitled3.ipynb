```python
class FiguraGeometrica:
    def __init__(self, alto, ancho):
        if not isinstance(alto, (int, float)) or alto <= 0:
            raise ValueError("El alto debe ser un número positivo.")
        if not isinstance(ancho, (int, float)) or ancho <= 0:
            raise ValueError("El ancho debe ser un número positivo.")
        self._alto = alto
        self._ancho = ancho

    def area(self):
        return self._alto * self._ancho

    def __str__(self):
        return f"Figura Geométrica: Alto={self._alto}, Ancho={self._ancho}"

# Ejemplo de uso de la clase FiguraGeometrica
figura1 = FiguraGeometrica(alto=10, ancho=5)
print(figura1)
print(f"Área: {figura1.area()}")

# Intentar crear una figura con valores inválidos (esto debería lanzar un error)
try:
    figura2 = FiguraGeometrica(alto=-2, ancho=5)
except ValueError as e:
    print(f"Error al crear figura2: {e}")

try:
    figura3 = FiguraGeometrica(alto=10, ancho="cinco")
except ValueError as e:
    print(f"Error al crear figura3: {e}")
```

```python
class Cuadrado(FiguraGeometrica):
    def __init__(self, lado):
        if not isinstance(lado, (int, float)) or lado <= 0:
            raise ValueError("El lado debe ser un número positivo.")
        super().__init__(alto=lado, ancho=lado)
        self._lado = lado

    def perimetro(self):
        return 4 * self._lado

    def __str__(self):
        return f"Cuadrado: Lado={self._lado}, Área={self.area()}, Perímetro={self.perimetro()}"

# Ejemplo de uso de la clase Cuadrado
cuadrado1 = Cuadrado(lado=7)
print(cuadrado1)
print(f"Área: {cuadrado1.area()}")
print(f"Perímetro: {cuadrado1.perimetro()}")

# Intentar crear un cuadrado con valores inválidos
try:
    cuadrado2 = Cuadrado(lado=-5)
except ValueError as e:
    print(f"Error al crear cuadrado2: {e}")

try:
    cuadrado3 = Cuadrado(lado="siete")
except ValueError as e:
    print(f"Error al crear cuadrado3: {e}")
```

```python
class Rectangulo(FiguraGeometrica):
    def __init__(self, ancho, alto):
        # Call the constructor of the parent class (FiguraGeometrica)
        super().__init__(alto=alto, ancho=ancho)
        self._ancho = ancho
        self._alto = alto

    def perimetro(self):
        return 2 * (self._ancho + self._alto)

    def __str__(self):
        return f"Rectángulo: Ancho={self._ancho}, Alto={self._alto}, Área={self.area()}, Perímetro={self.perimetro()}"

# Ejemplo de uso de la clase Rectangulo
rectangulo1 = Rectangulo(ancho=8, alto=4)
print(rectangulo1)
print(f"Área: {rectangulo1.area()}")
print(f"Perímetro: {rectangulo1.perimetro()}")

# Intentar crear un rectángulo con valores inválidos
try:
    rectangulo2 = Rectangulo(ancho=-3, alto=6)
except ValueError as e:
    print(f"Error al crear rectangulo2: {e}")

try:
    rectangulo3 = Rectangulo(ancho=10, alto="tres")
except ValueError as e:
    print(f"Error al crear rectangulo3: {e}")
```

```python
def sumar_areas(figuras):
    total = 0
    for figura in figuras:
        total += figura.area()
    return total


def sumar_perimetros(figuras):
    total = 0
    for figura in figuras:
        total += figura.perimetro()
    return total


try:
    cuadrado1 = Cuadrado(5)
    cuadrado2 = Cuadrado(8)

    rectangulo1 = Rectangulo(4, 6)
    rectangulo2 = Rectangulo(3, 7)

    figuras = [cuadrado1, cuadrado2, rectangulo1, rectangulo2]

    for figura in figuras:
        print(figura)
        print("Área:", figura.area())
        print("Perímetro:", figura.perimetro())
        print("-" * 30)

    print("Suma total de áreas:", sumar_areas(figuras))
    print("Suma total de perímetros:", sumar_perimetros(figuras))

    print("\nModificando valores...")
    rectangulo1.ancho = 10
    rectangulo1.alto = 2
    print(rectangulo1)

    print("\nProbando error:")
    cuadrado1.ancho = -3

except ValueError as e:
    print("Error:", e)
```
